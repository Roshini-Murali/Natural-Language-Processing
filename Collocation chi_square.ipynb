```python

import nltk
from nltk.corpus import stopwords
from nltk.util import bigrams
from collections import Counter
import math
import string
critical=2.576
f=open('Article.txt','r')
text=f.read()
text=text.split("\n")
print(text)

# List to store words that are tokenized
words=[]
for i in text:
    i=''.join(char for char in i if char not in string.punctuation)
    sample=nltk.word_tokenize(i)
    words.append(sample)
print(words)

# Concatenating lists
new=[]
for i in words:
  new+=i
print(new)

# List to store words by removing stopwords
main=[]
for w in new:
  if w not in set(stopwords.words('english')):
      main.append(w)
print(main)

unique_words=list(set(main))
N=len(main)

each_word=Counter(main)
print("\n",each_word) # count of each word

bigram_list=list(bigrams(main))
print("\nThe length of bigram list: ",len(bigram_list))
bigram_freq=Counter(bigram_list)
print("\n",bigram_freq) # count of bigrams

for i in bigram_freq.keys():

  f_bigram=bigram_freq[i] # frequency of the current bigram
  n1=each_word[i[0]] # frequency of the first word in the current bigram
  n2=each_word[i[1]] # frequency of the second word in the current bigram
  print("\n",f_bigram," ",n1," ",n2)

  # Observed frequency matrix m1
  m1=[f_bigram,n1-f_bigram,n2-f_bigram,N-n1-n2-f_bigram]

  # Expected frequency matrix m2
  m2=[((m1[0]+m1[1])*(m1[0]+m1[2]))/N, ((m1[0]+m1[1])*(m1[1]+m1[3]))/N, ((m1[2]+m1[3])*(m1[0]+m1[2]))/N, ((m1[3]+m1[1])*(m1[3]+m1[2]))/N]

  chi_square=0
  for k in range(4):
    chi_square+=(((m1[k]-m2[k])**2)/m2[k]) # Calculating chi square

  print("The result is ",chi_square)

  if chi_square > critical:
    print(i,"is a collocation")
  else:
    print(i,"is not a collocation")
```
